```
namespace enums;

public enum Color
{
    Red,
    Green,
    Blue
}
```

```
Console.WriteLine(Color.Red);
Console.WriteLine((int)Color.Red);
```

```
// Loop through all values
foreach (var c in Enum.GetValues<Color>())
{
    Console.WriteLine($"{c} = {(int)c}");
}
```

```

string Describe(Color c)
{
    return c switch
    {
        Color.Red => "Warm and energetic",
        Color.Green => "Calm and natural",
        Color.Blue => "Cool and stable",
        _ => "Unknown"
    };
}

Console.WriteLine(Describe(Color.Red));
```
